```python
import numpy as np
import pandas as pd
from matplotlib import pyplot as plt
```

```python
from do3se_phenology.phyllochron_dvi import calc_dvi
```

```python
td_full = np.arange(365) * 2
dd_full = np.arange(365)
hr_full = np.zeros(365)
sowing_day = 100
```

```python
td_at_sgs = 10
tt_emr = 10
tt_veg = 200
tt_rep = 300
dvi_x = [td_at_sgs,td_at_sgs + tt_emr, td_at_sgs+tt_emr+tt_veg, td_at_sgs + tt_emr + tt_veg + tt_rep]
dvi_y = [-1,0,1,2]

plt.plot(dvi_x, dvi_y, label="DVI", color="yellow")
```

```python
dvi = [-1]
rpe = np.ones(365)
teff = np.ones(365) * 18
tt_emr, tt_veg, tt_rep = [35, 1000, 666]
sowing_day = 100
for t, rp, d, hr in zip(teff, rpe, dd_full, hr_full):
  if hr == 0:
    new_dvi = calc_dvi(dvi[-1],t, tt_emr, tt_veg, tt_rep, rp,d, sowing_day)
    dvi.append(new_dvi)
plt.xlim(sowing_day - 5, sowing_day+120)
plt.plot(dvi)
plt.xlabel("Days")
plt.ylabel("DVI")
plt.show()
```

For each day we need to establish the number of emerged leaf populations

```python
from math import ceil
```

```python
nP = 3
leaf_pop_emerged = [ceil((i)*nP) if 0< i < 1 else nP if 0 < i else 0 for i in dvi]
plt.plot(dvi)
plt.plot(leaf_pop_emerged)
```
